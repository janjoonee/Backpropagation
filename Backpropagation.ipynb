```python

import numpy as np

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid Function
def sigmoid_derivative(x):
    return x * (1 - x)

# Inputs
i1, i2 = 0.05, 0.1

# Target Outputs
t_o1, t_o2 = 0.01, 0.99

# Initial Weights
w1, w2, w3, w4 = 0.15, 0.2, 0.25, 0.3
w5, w6, w7, w8 = 0.4, 0.45, 0.5, 0.55

# Biases
b1, b2 = 0.35, 0.6

# Learning Rate
alpha = 0.5

# Forward Pass
h1_in = i1 * w1 + i2 * w2 + b1
h1 = sigmoid(h1_in)

h2_in = i1 * w3 + i2 * w4 + b1
h2 = sigmoid(h2_in)

o1_in = h1 * w5 + h2 * w6 + b2
o1 = sigmoid(o1_in)

o2_in = h1 * w7 + h2 * w8 + b2
o2 = sigmoid(o2_in)

# Backward Pass
# Output Layer Error
delta_o1 = (o1 - t_o1) * sigmoid_derivative(o1)
delta_o2 = (o2 - t_o2) * sigmoid_derivative(o2)

# Output Layer Weight Updates
delta_w5 = h1 * delta_o1
w5 = w5 - alpha * delta_w5

delta_w6 = h2 * delta_o1
w6 = w6 - alpha * delta_w6

delta_w7 = h1 * delta_o2
w7 = w7 - alpha * delta_w7

delta_w8 = h2 * delta_o2
w8 = w8 - alpha * delta_w8

# Hidden Layer Error
delta_h1 = h1 * (1 - h1) * (delta_o1 * w5 + delta_o2 * w7)
delta_h2 = h2 * (1 - h2) * (delta_o1 * w6 + delta_o2 * w8)

# Hidden Layer Weight Updates
delta_w1 = i1 * delta_h1
w1 = w1 - alpha * delta_w1

delta_w2 = i2 * delta_h1
w2 = w2 - alpha * delta_w2

delta_w3 = i1 * delta_h2
w3 = w3 - alpha * delta_w3

delta_w4 = i2 * delta_h2
w4 = w4 - alpha * delta_w4

# Print Final Weights
print("Final Weights:")
print(f"w1: {w1:.4f}, w2: {w2:.4f}, w3: {w3:.4f}, w4: {w4:.4f}")
print(f"w5: {w5:.4f}, w6: {w6:.4f}, w7: {w7:.4f}, w8: {w8:.4f}")
```

```python

import numpy as np

# Input Layer Values
input_1, input_2 = 0.05, 0.10

# Target Outputs
target_1, target_2 = 0.01, 0.99

# Learning Rate
learning_rate = 0.5

# Initial Weights and Biases
weights = {
    'w1': 0.15, 'w2': 0.20, 'w3': 0.25, 'w4': 0.30,
    'w5': 0.40, 'w6': 0.45, 'w7': 0.50, 'w8': 0.55
}

bias = {
    'b1': 0.35,
    'b2': 0.60
}

# Activation Function (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid Function
def sigmoid_derivative(x):
    return x * (1 - x)

# Forward Pass
hidden_1_input = input_1 * weights['w1'] + input_2 * weights['w2'] + bias['b1']
hidden_1_output = sigmoid(hidden_1_input)

hidden_2_input = input_1 * weights['w3'] + input_2 * weights['w4'] + bias['b1']
hidden_2_output = sigmoid(hidden_2_input)

output_1_input = hidden_1_output * weights['w5'] + hidden_2_output * weights['w6'] + bias['b2']
output_1_output = sigmoid(output_1_input)

output_2_input = hidden_1_output * weights['w7'] + hidden_2_output * weights['w8'] + bias['b2']
output_2_output = sigmoid(output_2_input)

# Backward Pass
error_output_1 = (output_1_output - target_1) * sigmoid_derivative(output_1_output)
error_output_2 = (output_2_output - target_2) * sigmoid_derivative(output_2_output)

# Adjust Output Layer Weights
weights['w5'] -= learning_rate * error_output_1 * hidden_1_output
weights['w6'] -= learning_rate * error_output_1 * hidden_2_output
weights['w7'] -= learning_rate * error_output_2 * hidden_1_output
weights['w8'] -= learning_rate * error_output_2 * hidden_2_output

# Hidden Layer Errors
error_hidden_1 = sigmoid_derivative(hidden_1_output) * (error_output_1 * weights['w5'] + error_output_2 * weights['w7'])
error_hidden_2 = sigmoid_derivative(hidden_2_output) * (error_output_1 * weights['w6'] + error_output_2 * weights['w8'])

# Adjust Hidden Layer Weights
weights['w1'] -= learning_rate * error_hidden_1 * input_1
weights['w2'] -= learning_rate * error_hidden_1 * input_2
weights['w3'] -= learning_rate * error_hidden_2 * input_1
weights['w4'] -= learning_rate * error_hidden_2 * input_2

# Display Final Weights
print("Updated Weights:")
for key, value in weights.items():
    print(f"{key}: {value:.4f}")
```
